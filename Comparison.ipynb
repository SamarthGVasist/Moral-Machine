```python
import pandas as pd
import numpy as np
```

```python
data_og = pd.read_csv('SharedResponsesSurvey.csv')
data_cc = pd.read_csv('mproject(2).csv')
```

```python
data_og = data_og[['ScenarioOrder','ScenarioTypeStrict', 'Saved','Review_age', 'Review_gender','UserCountry3']]
data_og = data_og.dropna()
data_og
```

### Selecting the age and gender scenarios

```python
#select only india
data_og_india = data_og[(data_og['UserCountry3'] == "IND")]
data_og_india_age = data_og_india[(data_og_india['ScenarioTypeStrict'] == "Age")]
data_og_india_gender= data_og_india[(data_og_india['ScenarioTypeStrict'] == "Gender")]
```

### Cleaning data

```python
data_cc = data_cc.replace(to_replace = 1, value = 0)
data_cc = data_cc.replace(to_replace = 2, value = 1)
data_cc
```

```python
#Gender Irrespective - age wise
data_cc_age = data_cc[["scene2"]]
```

```python
data_cc_age
```

# Checking difference of choices in the age scenario

```python
data_cc_age_old =  data_cc[(data_cc['scene2'] == 0)]
data_cc_age_young =  data_cc[(data_cc['scene2'] == 1)]
```

### Distribution from gathered data
-> older people ~ 26%

-> younger people ~ 74%


```python
x=(len(data_cc_age_old)/len(data_cc))*100
y=(len(data_cc_age_young)/len(data_cc))*100
print("proportion choosing to save older people:",x)
print("proportion choosing to save younger people:",y)
```

```python
#plotting
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = ['old', 'young']
students = [x,y]
ax.bar(langs,students)
plt.show()
```

### Distribution from MIT data
-> older people ~ 47%

-> younger people ~ 53%

```python
#data_og_india_age = data_og_india[(data_og_india['ScenarioTypeStrict'] == "Age")]
data_og_age_young =  data_og_india_age[(data_og_india_age['Saved'] == 0.0)]
data_og_age_old =  data_og_india_age[(data_og_india_age['Saved'] == 1.0)]
x=(len(data_og_age_old)/len(data_og_india_age))*100
y=(len(data_og_age_young)/len(data_og_india_age))*100
print("proportion choosing to save older people:",x)
print("proportion choosing to save younger people:",y)
```

```python
#plotting
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = ['old', 'young']
students = [x,y]
ax.bar(langs,students)
plt.show()
```

### Thus we have difference in age choices.
From the collected data, we see that the distribution favours selection of young.
From the gathered data, we see that the distribution is more balanced.
